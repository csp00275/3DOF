```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 링크 길이
l1, l2, l3 = 1, 1, 0.2

# 목표 위치 및 방향
x_target = 0.5
y_target = 0.8
phi_target = np.deg2rad(-90)

# 관절 제한 (도 단위)
limits_deg = [(-180, 180), (-150, 150), (-120, 120)]

def forward_kinematics_3dof(theta1, theta2, theta3):
    x0, y0 = 0, 0
    x1 = l1 * np.cos(theta1)
    y1 = l1 * np.sin(theta1)
    x2 = x1 + l2 * np.cos(theta1 + theta2)
    y2 = y1 + l2 * np.sin(theta1 + theta2)
    x3 = x2 + l3 * np.cos(theta1 + theta2 + theta3)
    y3 = y2 + l3 * np.sin(theta1 + theta2 + theta3)
    return [(x0, y0), (x1, y1), (x2, y2), (x3, y3)]

def solve_ik_3dof(x_target, y_target, phi_target):
    x_wrist = x_target - l3 * np.cos(phi_target)
    y_wrist = y_target - l3 * np.sin(phi_target)

    r_square = x_wrist**2 + y_wrist**2
    cos_theta2 = (r_square - l1**2 - l2**2) / (2 * l1 * l2)

    if abs(cos_theta2) > 1.0:
        return []

    solutions = []
    for theta2 in [-np.arccos(cos_theta2), np.arccos(cos_theta2)]:
        k1 = l1 + l2 * np.cos(theta2)
        k2 = l2 * np.sin(theta2)
        theta1 = np.arctan2(y_wrist, x_wrist) - np.arctan2(k2, k1)
        theta3 = phi_target - theta1 - theta2

        angles = [theta1, theta2, theta3]
        angles_deg = [np.rad2deg(a) for a in angles]

        valid = all(lo <= a <= hi for a, (lo, hi) in zip(angles_deg, limits_deg))
        if valid:
            solutions.append((angles_deg, forward_kinematics_3dof(theta1, theta2, theta3)))

    return solutions

# 역기구학 해 계산
solutions = solve_ik_3dof(x_target, y_target, phi_target)

# 시각화
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_xlim(-2.5, 2.5)
ax.set_ylim(-2.5, 2.5)
ax.set_aspect('equal')
ax.set_title("3자유도 역기구학 (Elbow-up / Elbow-down 해 비교)")

colors = ['blue', 'green']
labels = ['Elbow-down', 'Elbow-up']

for i, (angles_deg, points) in enumerate(solutions):
    xs, ys = zip(*points)
    ax.plot(xs, ys, 'o-', lw=2, color=colors[i], label=f"{labels[i]}:\nθ₁={angles_deg[0]:.1f}° θ₂={angles_deg[1]:.1f}° θ₃={angles_deg[2]:.1f}°")

    # 각도 라벨 텍스트
    for j, (x, y) in enumerate(points):
        if j == 1:
            text = f"θ₁={angles_deg[0]:.1f}°"
        elif j == 2:
            text = f"θ₂={angles_deg[1]:.1f}°"
        elif j == 3:
            text = f"θ₃={angles_deg[2]:.1f}°"
        else:
            text = "Base"
        ax.text(x + 0.1, y + 0.1, text, fontsize=9, ha='center', va='center', color=colors[i])



ax.plot(x_target, y_target, 'rx', markersize=10, label='목표 위치')
ax.legend(loc='upper right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import PillowWriter

# 링크 길이 및 제한
l1, l2, l3 = 1.0, 1.0, 0.2

# 정기구학 함수
def forward_kinematics_3dof(theta1, theta2, theta3):
    x0, y0 = 0, 0
    x1 = l1 * np.cos(theta1); y1 = l1 * np.sin(theta1)
    x2 = x1 + l2 * np.cos(theta1 + theta2); y2 = y1 + l2 * np.sin(theta1 + theta2)
    x3 = x2 + l3 * np.cos(theta1 + theta2 + theta3); y3 = y2 + l3 * np.sin(theta1 + theta2 + theta3)
    return [(x0, y0), (x1, y1), (x2, y2), (x3, y3)]

# 역기구학 함수
def solve_ik(x_target, y_target, phi_target):
    x_wrist = x_target - l3 * np.cos(phi_target)
    y_wrist = y_target - l3 * np.sin(phi_target)
    r2 = x_wrist**2 + y_wrist**2
    cos_t2 = (r2 - l1**2 - l2**2) / (2 * l1 * l2)
    if abs(cos_t2) > 1.0:
        return None
    for t2 in [-np.arccos(cos_t2), np.arccos(cos_t2)]:
        k1 = l1 + l2 * np.cos(t2)
        k2 = l2 * np.sin(t2)
        t1 = np.arctan2(y_wrist, x_wrist) - np.arctan2(k2, k1)
        t3 = phi_target - t1 - t2
        degs = np.rad2deg([t1, t2, t3])
        if all(-180 <= d <= 180 for d in degs):
            return [t1, t2, t3]
    return None

# Trajectory: φ = -90°
phi_neg90 = -np.pi/2
x_fixed = 0.5
y_vals = np.linspace(2.0, 0.5, 100)
trajectory = [(x_fixed, y, phi_neg90) for y in y_vals if solve_ik(x_fixed, y, phi_neg90) is not None]

# 워크스페이스 그리드 & 마스크
grid_res = 200
x = np.linspace(-2.5, 2.5, grid_res)
y = np.linspace(-2.5, 2.5, grid_res)
X, Y = np.meshgrid(x, y)

# 이론적 워크스페이스 (회색)
R = np.sqrt(X**2 + Y**2)
L_total = l1 + l2 + l3
L_min = abs(l1 - l2 - l3)
workspace_mask = (R >= L_min) & (R <= L_total)

# φ = -90° 워크스페이스 (연파랑)
phi_mask = np.zeros_like(X, dtype=bool)
for i in range(grid_res):
    for j in range(grid_res):
        if workspace_mask[i, j] and solve_ik(X[i, j], Y[i, j], phi_neg90) is not None:
            phi_mask[i, j] = True

# 시각화 설정
fig, ax = plt.subplots(figsize=(8, 4))
ax.contourf(X, Y, workspace_mask, levels=[0.5, 1], colors=['lightgray'], alpha=0.3)
ax.contourf(X, Y, phi_mask, levels=[0.5, 1], colors=['skyblue'], alpha=0.5)

line, = ax.plot([], [], 'o-', lw=2, color='blue')
target_dot, = ax.plot([], [], 'rx', markersize=8)
angle_text = ax.text(0.05, 0.95, '', transform=ax.transAxes, fontsize=9, va='top',
                     bbox=dict(facecolor='white', alpha=0.7))

# 궤적 선
traj_xs = [pt[0] for pt in trajectory]
traj_ys = [pt[1] for pt in trajectory]
ax.plot(traj_xs, traj_ys, '--', color='gray', linewidth=1, alpha=0.6)

ax.set_xlim(-2.2, 2.2); ax.set_ylim(0, 2.2); ax.set_aspect('equal')
ax.set_title("3DOF Workspace + φ = -90° Simulation")

# 애니메이션 초기화
def init():
    line.set_data([], []); target_dot.set_data([], []); angle_text.set_text('')
    return line, target_dot, angle_text

# 프레임별 애니메이션
def animate(i):
    x, y, phi = trajectory[i]
    angles = solve_ik(x, y, phi)
    theta1, theta2, theta3 = angles
    pts = forward_kinematics_3dof(theta1, theta2, theta3)
    xs, ys = zip(*pts)
    line.set_data(xs, ys)
    target_dot.set_data([x], [y])
    degs = np.rad2deg([theta1, theta2, theta3])
    angle_text.set_text(f"θ₁={degs[0]:.1f}°  θ₂={degs[1]:.1f}°  θ₃={degs[2]:.1f}°")
    return line, target_dot, angle_text

ani = animation.FuncAnimation(fig, animate, frames=len(trajectory),
                              init_func=init, interval=50, blit=False)

# GIF 저장
ani.save("vertical_trajectory_linear.gif", writer=PillowWriter(fps=20))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import PillowWriter

# 링크 길이 및 각도 제한
l1, l2, l3 = 1.0, 1.0, 0.2
limits_deg = [(-180, 180), (-180, 180), (-180, 180)]

# 순기구학
def forward_kinematics_3dof(th1, th2, th3):
    x0, y0 = 0, 0
    x1, y1 = l1*np.cos(th1), l1*np.sin(th1)
    x2 = x1 + l2*np.cos(th1 + th2);  y2 = y1 + l2*np.sin(th1 + th2)
    x3, y3 = x2 + l3*np.cos(th1 + th2 + th3), y2 + l3*np.sin(th1 + th2 + th3)
    return [(x0, y0), (x1, y1), (x2, y2), (x3, y3)]

# 역기구학 (각도 제한 포함)
def solve_ik(x_t, y_t, phi):
    x_w = x_t - l3*np.cos(phi)
    y_w = y_t - l3*np.sin(phi)
    r2 = x_w**2 + y_w**2
    c2 = (r2 - l1**2 - l2**2)/(2*l1*l2)
    if abs(c2) > 1.0:
        return None
    for th2 in [-np.arccos(c2), np.arccos(c2)]:
        k1, k2 = l1 + l2*np.cos(th2), l2*np.sin(th2)
        th1 = np.arctan2(y_w, x_w) - np.arctan2(k2, k1)
        th3 = phi - th1 - th2
        degs = np.rad2deg([th1, th2, th3])
        if all(lo <= d <= hi for d,(lo,hi) in zip(degs, limits_deg)):
            return [th1, th2, th3]
    return None

# 워크스페이스 배경 계산
grid = 300
xg = np.linspace(-2.5, 2.5, grid)
yg = np.linspace(0, 2.5, grid)
Xg, Yg = np.meshgrid(xg, yg)
R = np.sqrt(Xg**2 + Yg**2)
L_total = l1 + l2 + l3
L_min = abs(l1 - l2 - l3)
workspace_mask = (R >= L_min) & (R <= L_total)

# φ = -90° 영역 마스크
phi = -np.pi/2
phi_mask = np.zeros_like(workspace_mask)
for i in range(grid):
    for j in range(grid):
        if workspace_mask[i,j] and solve_ik(Xg[i,j], Yg[i,j], phi):
            phi_mask[i,j] = True


# 수직 이동 궤적
x_fixed = 0.5
y0, y1 = 1.5, 0.5        # 시작/끝 위치
D = abs(y1 - y0)         # 전체 이동 거리
Vmax = 0.5               # 최대 속도
a_max = 0.2              # 최대 가속도

t_acc = Vmax / a_max
d_acc = 0.5 * a_max * t_acc**2

if 2 * d_acc >= D:
    # 삼각형 프로파일 (Vmax 미도달)
    t_acc = np.sqrt(D / a_max)
    t_flat = 0.0
else:
    # 트래피즈 프로파일
    t_flat = (D - 2 * d_acc) / Vmax

# 전체 시간
T = 2 * t_acc + t_flat

# 시간 벡터
N = 200
t = np.linspace(0, T, N)

# s(t) (이동 거리)
s = np.zeros_like(t)
for i, ti in enumerate(t):
    if ti <= t_acc:
        s[i] = 0.5 * a_max * ti**2
    elif ti <= t_acc + t_flat:
        s[i] = d_acc + Vmax * (ti - t_acc)
    else:
        td = ti - t_acc - t_flat
        s[i] = d_acc + Vmax * t_flat + Vmax * td - 0.5 * a_max * td**2

# s_norm: 0→1 정규화
s_norm = s / D


# 클램핑: 워크스페이스 φ-mask 범위
ix = np.argmin(np.abs(xg - x_fixed))
valid_ys = yg[phi_mask[:,ix]]
if valid_ys.size:
    y_min, y_max = valid_ys.min(), valid_ys.max()
    y_vals = np.clip(y_vals, y_min, y_max)

trajectory = [(x_fixed, y_vals[i], phi) for i in range(N)]

# y 속도, 가속도
dt = t[1] - t[0]
vel_y = np.gradient(y_vals, dt)
acc_y = np.gradient(vel_y, dt)

# 애니메이션 및 그래프 설정
fig = plt.figure(figsize=(10,6))
gs = fig.add_gridspec(3,2, width_ratios=[1,1], height_ratios=[1,1,1], wspace=0.4, hspace=0.4)
ax_robot = fig.add_subplot(gs[:,0])
ax_pos   = fig.add_subplot(gs[0,1])
ax_vel   = fig.add_subplot(gs[1,1])
ax_acc   = fig.add_subplot(gs[2,1])

# 로봇 플롯 (왼쪽)
ax_robot.contourf(Xg, Yg, workspace_mask, levels=[0.5,1], colors=['lightgray'], alpha=0.3)
ax_robot.contourf(Xg, Yg, phi_mask, levels=[0.5,1], colors=['skyblue'], alpha=0.5)
# 궤적 회색 점선
ax_robot.plot([x_fixed]*N, y_vals, '--', color='gray', linewidth=1, alpha=0.7)
line, = ax_robot.plot([], [], 'o-', lw=2, color='blue')
dot,  = ax_robot.plot([], [], 'rx', markersize=6)
ax_robot.set_xlim(-2.2,2.2); ax_robot.set_ylim(0,2.2); ax_robot.set_aspect('equal')
ax_robot.set_title("Workspace & Trajectory (φ=-90°)")

# 그래프 플롯 (오른쪽)
ax_pos.plot(t, y_vals, '-', lw=2);    ax_pos.set_ylabel("y Pos"); ax_pos.set_xlim(0, T)
ax_vel.plot(t, vel_y, '-', lw=2);     ax_vel.set_ylabel("y Vel"); ax_vel.set_xlim(0, T)
ax_acc.plot(t, acc_y, '-', lw=2);     ax_acc.set_ylabel("y Acc"); ax_acc.set_xlim(0, T)
for ax in (ax_pos, ax_vel, ax_acc):
    ax.set_xlabel("Time (s)")

marker_pos, = ax_pos.plot([], [], 'ro')
marker_vel, = ax_vel.plot([], [], 'ro')
marker_acc, = ax_acc.plot([], [], 'ro')

# 초기화 및 애니메이션 함수
def init():
    line.set_data([], []); dot.set_data([], [])
    marker_pos.set_data([], []); marker_vel.set_data([], []); marker_acc.set_data([], [])
    return line, dot, marker_pos, marker_vel, marker_acc

def animate(i):
    x,y,p = trajectory[i]
    th = solve_ik(x, y, p)
    if th:
        pts = forward_kinematics_3dof(*th)
        xs, ys = zip(*pts)
        line.set_data(xs, ys)
        dot.set_data([x],[y])
    marker_pos.set_data([t[i]],[y_vals[i]])
    marker_vel.set_data([t[i]],[vel_y[i]])
    marker_acc.set_data([t[i]],[acc_y[i]])
    return line, dot, marker_pos, marker_vel, marker_acc

ani = animation.FuncAnimation(fig, animate, frames=N, init_func=init, interval=50, blit=False)
ani.save("trap_with_graphs.gif", writer=PillowWriter(fps=20))

print("GIF saved: /mnt/data/trap_with_graphs.gif")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import PillowWriter

# 링크 길이 및 관절 제한
l1, l2, l3 = 1.0, 1.0, 0.2
limits_deg = [(-180,180),(-180,180),(-180,180)]

# 순기구학
def forward_kinematics_3dof(th1, th2, th3):
    x0,y0 = 0,0
    x1,y1 = l1*np.cos(th1), l1*np.sin(th1)
    x2,y2 = x1 + l2*np.cos(th1+th2), y1 + l2*np.sin(th1+th2)
    x3,y3 = x2 + l3*np.cos(th1+th2+th3), y2 + l3*np.sin(th1+th2+th3)
    return [(x0,y0),(x1,y1),(x2,y2),(x3,y3)]

# 역기구학 (제한 포함)
def solve_ik(x_t,y_t,phi):
    x_w = x_t - l3*np.cos(phi)
    y_w = y_t - l3*np.sin(phi)
    r2 = x_w**2 + y_w**2
    c2 = (r2 - l1**2 - l2**2)/(2*l1*l2)
    if abs(c2)>1.0: return None
    for th2 in [-np.arccos(c2), np.arccos(c2)]:
        k1,k2 = l1 + l2*np.cos(th2), l2*np.sin(th2)
        th1 = np.arctan2(y_w,x_w) - np.arctan2(k2,k1)
        th3 = phi - th1 - th2
        degs = np.rad2deg([th1,th2,th3])
        if all(lo<=d<=hi for d,(lo,hi) in zip(degs,limits_deg)):
            return [th1,th2,th3]
    return None

# 배경 워크스페이스 마스크
grid = 200
xg = np.linspace(-2.5,2.5,grid)
yg = np.linspace(0,2.5,grid)
Xg,Yg = np.meshgrid(xg,yg)
R = np.sqrt(Xg**2 + Yg**2)
Ltot = l1+l2+l3; Lmin = abs(l1-l2-l3)
ws_mask = (R>=Lmin)&(R<=Ltot)

# φ=0° 영역 (빨강)
phi0_mask = np.zeros_like(ws_mask)
for i in range(grid):
    for j in range(grid):
        if ws_mask[i,j] and solve_ik(Xg[i,j],Yg[i,j],0) is not None:
            phi0_mask[i,j] = True

# 트래젝토리: y_fixed=0.5, x 이동 길이=1.5
y_fixed = 0.5
x0,x1 = 2, 0.5
# 사다리꼴 프로파일 거리 파라미터
D = abs(x0 - x1)
Vmax = 0.3
a_max = 0.2
t_acc = Vmax/a_max
d_acc = 0.5*a_max*t_acc**2
if 2*d_acc >= D:
    t_acc = np.sqrt(D/a_max)
    t_flat = 0.0
else:
    t_flat = (D - 2*d_acc)/Vmax
T = 2*t_acc + t_flat

N = 200
t = np.linspace(0, T, N)
s = np.zeros_like(t)
for k, ti in enumerate(t):
    if ti <= t_acc:
        s[k] = 0.5 * a_max * ti**2
    elif ti <= t_acc + t_flat:
        s[k] = d_acc + Vmax*(ti - t_acc)
    else:
        td = ti - t_acc - t_flat
        s[k] = d_acc + Vmax*t_flat + Vmax*td - 0.5*a_max*td**2
x_vals = x_min + s/D * (x_max - x_min)

# 속도, 가속도 계산
dt = t[1] - t[0]
vel_x = np.gradient(x_vals, dt)
acc_x = np.gradient(vel_x, dt)

# 애니메이션 + 그래프
fig = plt.figure(figsize=(10,6))
gs = fig.add_gridspec(3,2, width_ratios=[1,1], height_ratios=[1,1,1], wspace=0.4, hspace=0.4)
ax_r = fig.add_subplot(gs[:,0])
ax_px = fig.add_subplot(gs[0,1])
ax_vx = fig.add_subplot(gs[1,1])
ax_ax = fig.add_subplot(gs[2,1])

# 로봇축 (좌) 그리기
ax_r.contourf(Xg,Yg,ws_mask,levels=[0.5,1],colors=['lightgray'],alpha=0.3)
ax_r.contourf(Xg,Yg,phi0_mask,levels=[0.5,1],colors=['salmon'],alpha=0.5)
ax_r.plot(x_vals, [y_fixed]*N, '--', color='gray', linewidth=1)
line, = ax_r.plot([], [], 'o-', lw=2, color='blue')
dot,  = ax_r.plot([], [], 'rx', markersize=6)
ax_r.set_xlim(-2.2,2.2); ax_r.set_ylim(0,2.2); ax_r.set_aspect('equal')

# 그래프 (우)
ax_px.plot(t, x_vals, '-', lw=2); ax_px.set_ylabel("x Position (m)"); ax_px.set_xlim(0, T)
ax_vx.plot(t, vel_x, '-', lw=2);    ax_vx.set_ylabel("x Velocity (m/s)"); ax_vx.set_xlim(0, T)
ax_ax.plot(t, acc_x, '-', lw=2);    ax_ax.set_ylabel("x Acceleration (m/s²)"); ax_ax.set_xlim(0, T)
for ax in (ax_px, ax_vx, ax_ax):
    ax.set_xlabel("Time (s)")


marker_pos, = ax_px.plot([], [], 'ro')
marker_vel, = ax_vx.plot([], [], 'ro')
marker_acc, = ax_ax.plot([], [], 'ro')

# 애니메이션 초기화
def init():
    line.set_data([], []); dot.set_data([], [])
    marker_pos.set_data([], []); marker_vel.set_data([], []); marker_acc.set_data([], [])
    return line, dot, marker_pos, marker_vel, marker_acc

def animate(i):
    x = x_vals[i]; y = y_fixed; p = 0.0
    th = solve_ik(x, y, p)
    if th:
        pts = forward_kinematics_3dof(*th)
        xs, ys = zip(*pts)
        line.set_data(xs, ys); dot.set_data([x],[y])
    marker_pos.set_data([t[i]],[x_vals[i]])
    marker_vel.set_data([t[i]],[vel_x[i]])
    marker_acc.set_data([t[i]],[acc_x[i]])
    return line, dot, marker_pos, marker_vel, marker_acc

ani = animation.FuncAnimation(fig, animate, frames=N, init_func=init, interval=50, blit=False)
ani.save("horizontal_trap_with_graphs.gif", writer=PillowWriter(fps=20))
print("GIF saved: horizontal_trap_with_graphs.gif")
```
